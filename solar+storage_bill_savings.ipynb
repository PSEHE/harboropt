```python
import pandas as pd
import os
import numpy as np

%reload_ext autoreload
%autoreload 2
```

```python
reopt_comm_bldgs = pd.read_csv('data/reopt_comm_bldgs.csv')
reopt_comm_bldgs['demand_charge_savings_per_battery_kw']=reopt_comm_bldgs['demand_charge_savings']/reopt_comm_bldgs['battery_kw']
```

```python
comm_floorspace_weights = pd.read_csv('data/cec_ladwp_comm_floorspace_weights.csv')
```

```python
for bldg_type in comm_floorspace_weights['building_type'].unique():
    if bldg_type != 'Miscellaneous':
        bldg_inds = reopt_comm_bldgs['Building']==bldg_type
        num_floors = reopt_comm_bldgs.loc[bldg_inds,'floors'].item()
        bldg_inds_floorspace = comm_floorspace_weights['building_type'] == bldg_type
        comm_floorspace_weights.loc[bldg_inds_floorspace,'roof_sqft'] = comm_floorspace_weights.loc[bldg_inds_floorspace,'floor_million_sqft']*1000000/num_floors
        
```

```python
total_roof_sqft = comm_floorspace_weights['roof_sqft'].sum()
```

```python
comm_floorspace_weights['weight'] = comm_floorspace_weights['roof_sqft']/total_roof_sqft
```

```python
for bldg_type in comm_floorspace_weights['building_type'].unique():
    if bldg_type != 'Miscellaneous':
        bldg_inds = reopt_comm_bldgs['Building']==bldg_type
        bldg_inds_floorspace = comm_floorspace_weights['building_type'] == bldg_type
        
        comm_floorspace_weights.loc[bldg_inds_floorspace,'energy_cost_savings_per_pv_kw_yearly'] = reopt_comm_bldgs.loc[bldg_inds,'energy_cost_savings_per_pv_kw'].item()
        comm_floorspace_weights.loc[bldg_inds_floorspace,'demand_charge_savings_per_batt_kw_yearly'] = reopt_comm_bldgs.loc[bldg_inds,'demand_charge_savings_per_battery_kw'].item()
 
        comm_floorspace_weights.loc[bldg_inds_floorspace,'pv_kw_to_battery_kw_ratio'] = reopt_comm_bldgs.loc[bldg_inds,'ratio_pv_kw_to_battery_kw'].item()
    
```
