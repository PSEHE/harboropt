```python
import pandas as pd
import numpy as np
```

```python
harborgen = pd.read_csv('data/EPADownload\\emission_01-10-2020_035150571.csv',index_col=False, skipinitialspace=True)
harborgen.fillna(0, inplace=True) 
harborgen['datetime'] = pd.to_datetime(harborgen['Date']) + pd.to_timedelta(harborgen['Hour'], unit='h')
harborgen.drop(['Date','Hour','Year'],inplace=True,axis=1)
harborgen.set_index(['datetime'])
```
